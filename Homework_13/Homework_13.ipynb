**Пункт 1.1 функции для подготовки (очищения) текста**

```python
def lower_сase(text):
  '''Приводит текст к нижнему регистру'''
  return text.lower()
```

```python
from nltk.corpus.reader.ycoe import RegexpTokenizer
def tokenized(text):
  '''Токенизирует текст при помощи регулярного выражения,
     убирает знаки пунктуации и убирает стоп слова'''
  tokeniz = RegexpTokenizer(r'\w{2,}')
  text1 = tokeniz.tokenize(text)
  tokens_without_sw = [word for word in text1 if not word in stopwords.words('russian')]
  return tokens_without_sw
```

```python
def lematizer(text):
  '''лематизирует текст'''
  lemmatized_words = [snowball.stem(word) for word in text]
  return lemmatized_words
```

```python
!pip install transformers
import pandas as pd
import torch
from torch.utils.data import TensorDataset, DataLoader, RandomSampler
from transformers import BertTokenizer, BertForSequenceClassification, AdamW, get_linear_schedule_with_warmup

# Загрузка датасета IMDb
url = '/content/drive/MyDrive/Dataset/IMDB-Dataset.csv'
df = pd.read_csv(url)

# Отображение меток на целые числа (0 для 'negative' и 1 для 'positive')
df['sentiment'] = df['sentiment'].map({'negative': 0, 'positive': 1})

# Загрузка предварительно обученного токенизатора BERT
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased', do_lower_case=True)

# Токенизация и предобработка текстовых данных
MAX_LEN = 128

def preprocess_text(text):
    inputs = tokenizer.encode_plus(
        text,
        add_special_tokens=True,
        max_length=MAX_LEN,
        padding='max_length',
        return_tensors='pt',
        return_attention_mask=True,
        truncation=True
    )
    return inputs['input_ids'], inputs['attention_mask']

# Применяем функцию preprocess_text к каждому тексту в датасете
input_ids, attention_mask = zip(*df['review'].apply(preprocess_text))

# Преобразование данных в PyTorch тензоры и создание DataLoader для обучающего набора
train_input_ids = torch.cat([torch.tensor(input_id) for input_id in input_ids])
train_attention_mask = torch.cat([torch.tensor(mask) for mask in attention_mask])
train_labels = torch.tensor(df['sentiment'].values)

batch_size = 16
train_data = TensorDataset(train_input_ids, train_attention_mask, train_labels)
train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)

# Загрузка предварительно обученной модели BERT для классификации последовательностей
model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=2)
model.cuda()

# Настройка оптимизатора и планировщика скорости обучения
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8)
epochs = 6
total_steps = len(train_dataloader) * epochs
scheduler = get_linear_schedule_with_warmup(optimizer, num_warmup_steps=0, num_training_steps=total_steps)

# Обучение модели
for epoch in range(epochs):
    model.train()
    total_loss = 0

    for batch in train_dataloader:
        batch = tuple(t.to('cuda') for t in batch)
        inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
        optimizer.zero_grad()
        outputs = model(**inputs)
        loss = outputs.loss
        loss.backward()
        optimizer.step()
        scheduler.step()
        total_loss += loss.item()

    avg_train_loss = total_loss / len(train_dataloader)
    print(f'Эпоха {epoch + 1}/{epochs}, Средний потери при обучении: {avg_train_loss:.4f}')

# Сохранение обученной модели
model.save_pretrained('/content/drive/MyDrive/Colab Notebooks/sentiment_model')

print('Обучение завершено!')
```

```python
!pip install pandas
!pip install transformers
!pip install torch
import pandas as pd
import torch
from transformers import BertTokenizer, BertForSequenceClassification
from transformers import AdamW

# Загрузка датасета
url = '/content/drive/MyDrive/Dataset/IMDB-Dataset.csv'
df = pd.read_csv(url)

# Вывод первых нескольких строк датасета для просмотра данных
df.head()

# Загрузка предварительно обученной модели BERT и токенизатора
model_name = "bert-base-uncased"  # Используем предварительно обученную модель BERT
tokenizer = BertTokenizer.from_pretrained(model_name)
model = BertForSequenceClassification.from_pretrained(model_name, num_labels=2)  # 2 класса: положительный и отрицательный

# Пример данных из датасета
data = [
    "One of the other reviewers has mentioned that ...",
    "A wonderful little production. <br /><br />The...",
    "I thought this was a wonderful way to spend ti...",
    "Basically there's a family where a little boy ...",
    "Petter Mattei's 'Love in the Time of Money' is..."
]

labels = [1, 1, 1, 0, 1]  # Пример меток для данных: 1 - положительный, 0 - отрицательный

# Преобразование текстовых данных в формат BERT
input_ids = []
attention_masks = []

for text in data:
    encoded_dict = tokenizer(
        text,
        add_special_tokens=True,  # Добавление токенов [CLS] и [SEP]
        max_length=64,  # Максимальная длина текста
        padding='max_length',  # Добавление паддинга до максимальной длины
        truncation=True,  # Обрезка текста, если он длиннее
        return_tensors='pt'  # Возвращение тензоров PyTorch
    )

    input_ids.append(encoded_dict['input_ids'])
    attention_masks.append(encoded_dict['attention_mask'])

input_ids = torch.cat(input_ids, dim=0)
attention_masks = torch.cat(attention_masks, dim=0)
labels = torch.tensor(labels)


from sklearn.model_selection import train_test_split

# Разделение данных на обучающую и тестовую выборки
train_inputs, test_inputs, train_masks, test_masks, train_labels, test_labels = train_test_split(
    input_ids, attention_masks, labels, test_size=0.2, random_state=42
)


# Создание DataLoader для обучения и тестирования
from torch.utils.data import TensorDataset, DataLoader, RandomSampler, SequentialSampler

batch_size = 32

train_data = TensorDataset(train_inputs, train_masks, train_labels)
train_sampler = RandomSampler(train_data)
train_dataloader = DataLoader(train_data, sampler=train_sampler, batch_size=batch_size)

test_data = TensorDataset(test_inputs, test_masks, test_labels)
test_sampler = SequentialSampler(test_data)
test_dataloader = DataLoader(test_data, sampler=test_sampler, batch_size=batch_size)

# Оптимизатор и функция потерь
optimizer = AdamW(model.parameters(), lr=2e-5)
loss_fn = torch.nn.CrossEntropyLoss()

# Обучение модели
num_epochs = 3

for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for batch in train_dataloader:
        input_ids, input_mask, labels = batch

        optimizer.zero_grad()
        outputs = model(input_ids, attention_mask=input_mask, labels=labels)
        loss = outputs.loss
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}, Loss: {total_loss}")

# Оценка модели на тестовых данных
model.eval()
test_loss = 0
test_correct = 0

with torch.no_grad():
    for batch in test_dataloader:
        input_ids, input_mask, labels = batch
        outputs = model(input_ids, attention_mask=input_mask, labels=labels)
        loss = outputs.loss
        logits = outputs.logits
        test_loss += loss.item()
        _, predicted = torch.max(logits, dim=1)
        test_correct += torch.sum(predicted == labels).item()

# Вывод метрик
print(f"Test Loss: {test_loss}")
print(f"Test Accuracy: {test_correct / len(test_data)}")
```
