Гипотеза №2. Модели из серии ноутбуков(NB) в среднем минимум на 10% тяжелее ультрабуков(UB) и модели из серии игровые(GM) в среднем минимум на 10% тяжелее моделей из серии ноутбуков(NB).

Задача: подтвердить корреляцию на графике

chart (1).png

Решение: выборка произведена на основе первых 10 моделей из каждой серии в датасете.

посчитаем средний вес каждой из серий с помощью python

```python
ub_weight = (1.37+1.34+1.83+1.87+2.04+1.34+2.2+1.3+1.6+1.83)/10
nb_weight = (1.86+2.1+1.86+1.86+2.2+2.2+2.2+1.86+2.3+2.2)/10
gm_weight = (2.5+2.65+3.2+2.43+3+2.7+2.2+2.65+2.2+2.45)/10
print('Средний вес ультрабуков равен', ub_weight)
print('Средний вес ноутбуков равен', nb_weight, 'и на', round((nb_weight/ub_weight)*100-100), 'процентов больше, чем средний вес ультрабуков')
print('Средний вес игровых ноутбуков равен', gm_weight, 'и на', round((gm_weight/nb_weight)*100-100), 'процентов больше, чем средний вес ноутбуков')
```

Вывод: гипотеза №2 подтверждена.
