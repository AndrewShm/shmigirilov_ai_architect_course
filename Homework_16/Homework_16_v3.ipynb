```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
df = pd.DataFrame(columns=['Primary Key', 'City', 'Temp', 'Precipitation', 'Time']) # создаем структуру базы данных
primary_key = 0
```

```python
df
```

```python
def weather_check(city):
  headers = {
      'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/116.0.5845.967 YaBrowser/23.9.1.967 Yowser/2.5 Safari/537.36'
  }

  res = requests.get(
        f'https://www.google.com/search?q={city}&oq={city}&aqs=chrome.0.35i39l2j0l4j46j69i60.6128j1j7&sourceid=chrome&ie=UTF-8',
        headers=headers
    )

  soup = BeautifulSoup(res.text, 'html.parser')

  time = soup.select('#wob_dts')[0].getText().strip()
  precipitation = soup.select('#wob_dc')[0].getText().strip()
  weather = soup.select('#wob_tm')[0].getText().strip()

  global primary_key
  global df
  primary_key = primary_key + 1

  new_row = {'Primary Key': primary_key, 'City': city.replace(' погода по цельсию', ''), 'Temp': weather, 'Precipitation': precipitation, 'Time': time}
  df = df.append(new_row, ignore_index = True)

  print(f'''День недели и время: {time}
Информация об осадках: {precipitation}
Температура воздуха: {weather}°C''')
```

```python
if __name__ == '__main__':
    city_input = input('Введите название города: ')
    weather_check(f'{city_input} погода по цельсию')
```
